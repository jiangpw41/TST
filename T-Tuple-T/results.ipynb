```python
from wayne_utils import load_data, save_data
import os
import pandas as pd

# 将字典转换为csv保存格式
def to_csv( points_count ):
    data = pd.DataFrame({
        'Team': ['Away Team', 'Home Team'],
        'Goals': [ None, None],
        'Shots': [None, None],
        'Fouls': [None, None],
        'Yellow Cards': [None, None],
        'Red Cards': [None, None],
        'Corner Kicks': [None, None],
        'Free Kicks': [None, None],
        'Offsides': [None, None]
    })
    list_ = "Goals,Shots,Fouls,Yellow Cards,Red Cards,Corner Kicks,Free Kicks,Offsides".split(",")
    for i in range( 8 ):
        data[ list_[i]][0] = points_count['away'][i]
        data[ list_[i]][1] = points_count['home'][i]
    return data
```

## 第一阶段结果

```python
# 后处理
points = ['goal', 'shot', 'foul', 'yellow card', 'red card', 'corner kick', 'free kick', 'offside']
map_ = {}
for i, key in enumerate(points):
    map_[key] = i
    
def single_doc_postprocess( predicts ):
    points_count = {
        "home": [ 0 for _ in range(8) ],
        "away": [ 0 for _ in range(8) ],
    }
    for tup in predicts:
        team_name = None
        field = None
        for item in tup:
            if "home team" in item.lower():
                team_name = "home"
            elif "away team" in item.lower():
                team_name = "away"
            else:
                for point in points:
                    if point in item.lower():
                        field = point
                        break
        if team_name == None or field == None:
            continue
        else:
            points_count[team_name][ map_[field] ] += 1
    return points_count
```

```python
# 后处理
points = ['goal', 'shot', 'foul', 'yellow card', 'red card', 'corner kick', 'free kick', 'offside']
map_ = {}
for i, key in enumerate(points):
    map_[key] = i

import re

def extract_scores(text):
    # 正则匹配两组 8 个逗号分隔的数字
    pattern = re.compile(r'(\d+,\d+,\d+,\d+,\d+,\d+,\d+,\d+)')
    matches = pattern.findall(text)

    # 调试信息：输出匹配到的结果
    print(f"Matches found: {matches}")
    
    # 如果找到的组数不等于 2，返回 None
    if len(matches) != 2:
        return None

    # 将匹配结果转为列表形式
    home_scores = list(map(int, matches[0].split(',')))
    away_scores = list(map(int, matches[1].split(',')))

    # 构建结果字典
    return {
        "home": home_scores,
        "away": away_scores
    }

def single_doc_postprocess( predicts ):
    points_count = {
        "home": [ 0 for _ in range(8) ],
        "away": [ 0 for _ in range(8) ],
    }
    if not isinstance( predicts, str):
        return points_count
    else:
        format = extract_scores(predicts)
        if format == None:
            return points_count
        else:
            return format
```
